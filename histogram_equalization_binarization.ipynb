```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load image
from google.colab import files
uploaded = files.upload()

# Specify the uploaded file name
img_path = list(uploaded.keys())[0]
img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)

# Histogram Equalization (HE)
he_img = cv2.equalizeHist(img)

# CLAHE (Contrast Limited AHE)
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
clahe_img = clahe.apply(img)

# Binarization (OTSU)
_, bin_img = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# Original & HE (2x2)
plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
plt.imshow(img, cmap='gray')
plt.title("Original Image")
plt.axis("off")

plt.subplot(2, 2, 2)
plt.hist(img.ravel(), 256, [0, 256])
plt.title("Original Histogram")

plt.subplot(2, 2, 3)
plt.imshow(he_img, cmap='gray')
plt.title("Histogram Equalization")
plt.axis("off")

plt.subplot(2, 2, 4)
plt.hist(he_img.ravel(), 256, [0, 256])
plt.title("HE Histogram")

plt.tight_layout()
plt.show()

# AHE (Adaptive Histogram Equalization)
ahe = cv2.createCLAHE(clipLimit=40.0, tileGridSize=(8, 8))
ahe_img = ahe.apply(img)

# Visualize AHE (2x2)
plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
plt.imshow(ahe_img, cmap='gray')
plt.title("AHE")
plt.axis("off")

plt.subplot(2, 2, 2)
plt.hist(ahe_img.ravel(), 256, [0, 256])
plt.title("AHE Histogram")

# CLAHE & OTSU (2x2)
plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
plt.imshow(clahe_img, cmap='gray')
plt.title("CLAHE")
plt.axis("off")

plt.subplot(2, 2, 2)
plt.hist(clahe_img.ravel(), 256, [0, 256])
plt.title("CLAHE Histogram")

plt.subplot(2, 2, 3)
plt.imshow(bin_img, cmap='gray')
plt.title("Binarization (OTSU)")
plt.axis("off")

plt.subplot(2, 2, 4)
plt.hist(bin_img.ravel(), 256, [0, 256])
plt.title("Binary Histogram")

plt.tight_layout()
plt.show()


```
