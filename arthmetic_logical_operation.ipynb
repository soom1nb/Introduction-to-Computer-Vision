# **addition**

```python
# Import necessary libraries
import cv2
from google.colab import files
import matplotlib.pyplot as plt

# Upload images (upload 2 files)
uploaded = files.upload()

# Check the uploaded filenames
filenames = list(uploaded.keys())
print("업로드된 파일:", filenames)

# Read the two images (in color)
img1 = cv2.imread(filenames[0])
img2 = cv2.imread(filenames[1])

# Resize the two images to the same size
h = min(img1.shape[0], img2.shape[0])
w = min(img1.shape[1], img2.shape[1])
img1_resized = cv2.resize(img1, (w, h))
img2_resized = cv2.resize(img2, (w, h))

# Addition
added = cv2.add(img1_resized, img2_resized)

# Visualization
plt.figure(figsize=(12,6))

plt.subplot(1,3,1)
plt.title("Image 1")
plt.imshow(cv2.cvtColor(img1_resized, cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.subplot(1,3,2)
plt.title("Image 2")
plt.imshow(cv2.cvtColor(img2_resized, cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.subplot(1,3,3)
plt.title("Added Image")
plt.imshow(cv2.cvtColor(added, cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.show()
```

```python
# Save the image using OpenCV
cv2.imwrite("added_image.jpg", added)
print("이미지 저장 완료: added_image.jpg")

# Provide a download link in Colab
from google.colab import files
files.download("added_image.jpg")
```

# **substraction**

```python
# Import necessary libraries
import cv2
from google.colab import files
import matplotlib.pyplot as plt

# Upload images (upload 2 files)
uploaded = files.upload()

# Check the uploaded filenames
filenames = list(uploaded.keys())
print("업로드된 파일:", filenames)

# Read the two images
img1 = cv2.imread(filenames[0])
img2 = cv2.imread(filenames[1])

# Resize the two images to the same size
h = min(img1.shape[0], img2.shape[0])
w = min(img1.shape[1], img2.shape[1])
img1_resized = cv2.resize(img1, (w, h))
img2_resized = cv2.resize(img2, (w, h))

# Subtraction
subtracted = cv2.subtract(img1_resized, img2_resized)

# Visualization
plt.figure(figsize=(12,6))

plt.subplot(1,3,1)
plt.title("Image 1")
plt.imshow(cv2.cvtColor(img1_resized, cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.subplot(1,3,2)
plt.title("Image 2")
plt.imshow(cv2.cvtColor(img2_resized, cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.subplot(1,3,3)
plt.title("Subtracted Image")
plt.imshow(cv2.cvtColor(subtracted, cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.show()
```

# **division**

```python
# Import necessary libraries
import cv2
from google.colab import files
import matplotlib.pyplot as plt
import numpy as np

# Upload images (upload 2 files)
uploaded = files.upload()

# Check the uploaded filenames
filenames = list(uploaded.keys())
print("업로드된 파일:", filenames)

# Read the two images
img1 = cv2.imread(filenames[0])
img2 = cv2.imread(filenames[1])

# Resize the two images to the same size
h = min(img1.shape[0], img2.shape[0])
w = min(img1.shape[1], img2.shape[1])
img1_resized = cv2.resize(img1, (w, h))
img2_resized = cv2.resize(img2, (w, h))

# Division
img2_safe = img2_resized.astype(np.float32) + 1e-5
divided = cv2.divide(img1_resized.astype(np.float32), img2_safe)

# Normalize the result to the 0-255 range and convert to uint8
divided = cv2.normalize(divided, None, 0, 255, cv2.NORM_MINMAX)
divided = divided.astype(np.uint8)

# Visualization
plt.figure(figsize=(12,6))

plt.subplot(1,3,1)
plt.title("Image 1")
plt.imshow(cv2.cvtColor(img1_resized, cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.subplot(1,3,2)
plt.title("Image 2")
plt.imshow(cv2.cvtColor(img2_resized, cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.subplot(1,3,3)
plt.title("Divided Image")
plt.imshow(cv2.cvtColor(divided, cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.show()
```

# **multiplication**

```python
# Import necessary libraries
import cv2
from google.colab import files
import matplotlib.pyplot as plt
import numpy as np

# Upload image (upload 1 file)
uploaded = files.upload()

# Check the uploaded filename
filename = list(uploaded.keys())[0]
print("업로드된 파일:", filename)

# Read the image
img = cv2.imread(filename)

#  Multiplication(itself)
img_mul_self = cv2.multiply(img, img)

# 결과는 값이 클 수 있어서 normalize
img_mul_self = cv2.normalize(img_mul_self, None, 0, 255, cv2.NORM_MINMAX)
img_mul_self = img_mul_self.astype(np.uint8)

# Multiplication(2x)
alpha = 2.0
img_mul_const = cv2.multiply(img.astype(np.float32), alpha)
img_mul_const = np.clip(img_mul_const, 0, 255).astype(np.uint8)

# Visualization
plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
plt.title("Original")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.subplot(1,3,2)
plt.title("Multiplication (Self * Self)")
plt.imshow(cv2.cvtColor(img_mul_self, cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.subplot(1,3,3)
plt.title(f"Multiplication (Const * {alpha})")
plt.imshow(cv2.cvtColor(img_mul_const, cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.show()
```

# **AND / OR / NOR**

```python
# Import necessary libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# Upload Images
uploaded = files.upload()
img_names = list(uploaded.keys())

# Load Images (Grayscale)
img1 = cv2.imread(img_names[0], cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread(img_names[1], cv2.IMREAD_GRAYSCALE)

# Resize to match sizes
img2_resized = cv2.resize(img2, (img1.shape[1], img1.shape[0]))

# Binarization
_, bin1 = cv2.threshold(img1, 127, 255, cv2.THRESH_BINARY)
_, bin2 = cv2.threshold(img2_resized, 127, 255, cv2.THRESH_BINARY)

# logical operation
and_result = cv2.bitwise_and(bin1, bin2)
or_result  = cv2.bitwise_or(bin1, bin2)
nor_result = cv2.bitwise_not(or_result)

# Visualization
plt.figure(figsize=(12, 12))

plt.subplot(3,2,1)
plt.title("Original Image 1")
plt.imshow(img1, cmap='gray')
plt.axis("off")

plt.subplot(3,2,2)
plt.title("Original Image 2")
plt.imshow(img2_resized, cmap='gray')
plt.axis("off")

plt.subplot(3,2,3)
plt.title("Binarized Image 1")
plt.imshow(bin1, cmap='gray')
plt.axis("off")

plt.subplot(3,2,4)
plt.title("Binarized Image 2")
plt.imshow(bin2, cmap='gray')
plt.axis("off")

plt.subplot(3,3,7)
plt.title("AND Result")
plt.imshow(and_result, cmap='gray')
plt.axis("off")

plt.subplot(3,3,8)
plt.title("OR Result")
plt.imshow(or_result, cmap='gray')
plt.axis("off")

plt.subplot(3,3,9)
plt.title("NOR Result")
plt.imshow(nor_result, cmap='gray')
plt.axis("off")

plt.tight_layout()
plt.show()
```
